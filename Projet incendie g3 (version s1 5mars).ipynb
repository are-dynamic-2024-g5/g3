```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
#coordonnée de la personne
x = 30
y = 50
pos = np.array((x,y))

#coordonnée de sortie
x0 = 0
y0 = 0
sortie = np.array((x0,y0))
direction = pos-sortie
direction = direction/np.sqrt(direction[0]**2+direction[1]**2) #vecteur unitaire
print(direction)

#vitesse de mouvement de la personne
v = 1.5 #m/s

#temps
t = np.arange(0,11,1)
```

```python
def dx(x,direction,v,t):
    l_x = []
    for i in (t):
        dx = v*i*direction[0]
        l_x.append(x-dx)
    return l_x

l_x = dx(x,direction,v,t)


def dy(y,direction,v,t):
    l_y = []
    for i in (t):
        dy = v*i*direction[1]
        l_y.append(y-dy)
    return l_y

l_y = dy(y,direction,v,t)


plt.plot(l_x,l_y)
plt.xlabel("x")
plt.ylabel("y")
plt.plot(x,y,"bo")
plt.plot(x0,y0,"ro")
plt.axis("equal")
plt.show()
```
